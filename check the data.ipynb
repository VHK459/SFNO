```python
def unstack_levels(ds):
    """
    Reshape (time, level, lon, lat) → (time, lon, lat)
    with each variable split into var_level0, var_level1, var_level2, etc.
    """
    new_vars = {}
    for var in ds.data_vars:
        if 'level' in ds[var].dims:
            for i, lev in enumerate(ds['level'].values):
                new_name = f"{var}_{lev}"
                new_vars[new_name] = ds[var].sel(level=lev).drop_vars('level')
        else:
            new_vars[var] = ds[var]

    return xr.Dataset(new_vars, attrs = ds.attrs)

vertical_flat_ds = unstack_levels(vertical_ds)
```

## checking nans

```python
%%time
ds = merged_ds
# 1. Identify only Time-Dependent Variables
# We skip static fields (like 'land_sea_mask') because they don't have a 'time' dimension
vars_with_time = [v for v in ds.data_vars if 'time' in ds[v].dims]

print(f"Checking {len(vars_with_time)} variables for NaNs...")

# 2. Build a Lazy Boolean Mask
# We don't calculate yet. We just build the plan.
overall_time_mask = None

for var_name in vars_with_time:
    # Find dimensions to reduce (e.g., ('latitude', 'longitude') or ('level', 'latitude', 'longitude'))
    dims_to_reduce = [d for d in ds[var_name].dims if d != 'time']
    
    # Check if ANY NaN exists in this variable for each time step
    # Result is a 1D DataArray of booleans along 'time'
    var_mask = ds[var_name].isnull().any(dim=dims_to_reduce)
    
    # Combine with previous masks using logical OR (|)
    if overall_time_mask is None:
        overall_time_mask = var_mask
    else:
        overall_time_mask = overall_time_mask | var_mask

# 3. Compute the Result
# This is the only step that triggers the computation
print("Computing NaN locations (this may take a moment)...")
overall_time_mask = overall_time_mask.compute()
```

```python
bad_times = ds.time[overall_time_mask]

print(f"Found {len(bad_times)} bad time steps.")
print(bad_times)
```

```python
ms_stacked_dataset = mg.to_stacked_array("channel", sample_dims=("time", "latitude", "longitude")).transpose( "channel","time", "latitude", "longitude")
```

```python
ms_stacked_dataset
names = [] 
for i in ms_stacked_dataset.indexes["channel"]:
    # Create name: "variable_level"
    print(i)
    new_name = str(i[0])
    names.append(new_name)

```

```python
ms_stacked_dataset_flat = ms_stacked_dataset.assign_coords(channel=names)
ms_stacked_dataset_flat
```

```python
ms_stacked_dataset_flat.name = "era5_features"
ms_stacked_dataset_flat.attrs = {
    "description": "ERA5 Stacked Data: 6 Vertical Levels + Surface Variables",
    "source": "ERA5 Reanalysis",
    "history": "Created via xarray.to_stacked_array"
}

# 2. CLEAR ENCODING (Crucial!)
# Xarray remembers the *original* chunk sizes in the 'encoding' attribute.
# If you don't clear this, it might try to revert to the old bad chunks when saving.
ms_stacked_dataset_flat.encoding = {} 

# 3. RE-CHUNK
# Goal: ~100MB per chunk.
# Calculation: 240 lat * 121 lon * 18 channels * 4 bytes = ~2 MB per time step.
# A time chunk of 48 gives: 48 * 2 MB = ~96 MB. This is perfect.
rechunked_da = ms_stacked_dataset_flat.chunk({
    "time": 48,         # 48 hours per block
    "latitude": -1,     # -1 means "don't split this dim" (keep full size)
    "longitude": -1,
    "channel": -1
})
import xarray as xr
import numcodecs

# ... (Previous steps: rechunking, to_dataset, etc.) ...
ds_to_save = rechunked_da.to_dataset()

# 1. Clear default encoding (as discussed before)
for var in ds_to_save.variables:
    ds_to_save[var].encoding = {}

# 2. THE FIX: Apply VLenUTF8 filter to the string coordinate
# This tells Zarr: "Don't use fixed 32-bit blocks. Use standard flexible strings."
ds_to_save['channel'] =  ds_to_save["channel"].astype(object)
# 3. Save
save_path = "/storage/vishnu/era5_1979_stacked.zarr"
print(f"Saving to {save_path}...")

ds_to_save.to_zarr(
    save_path, 
    mode="w", 
    safe_chunks=False
)
```
